### Signal Processing
#### Fourier series:
$x(t)=\sum_{n=-\infty}^{\infty}c_n e^{jn\omega_0t}=\sum_{n=-\infty}^{\infty}c_n e^{j\frac{2\pi n}{T}t} , \; \omega_0=\frac{2\pi}{T}$
<br><br> such that $x(t)=x(t+T)$

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
# a sawtooth wave with period=2pi
def sawtooth(t):
    return np.remainder((t-np.pi),2*np.pi)

# approximating the sawtooth with Foerier series
def sawtooth_FS(t,noterms=2):
    partialsum=0
    thesign=1
    for i in range(1,noterms+1):
        partialsum+=thesign*np.sin(i*t)/i
        thesign=-thesign
    return np.pi+2*partialsum
terms=['$2sint$','$-sin2t$','$+\\frac{2}{3}sin3t$',
       '$-\\frac{1}{2}sin4t$']
# demo
ts=np.linspace(-5*np.pi,5*np.pi,1000)
plt.figure(figsize=(12,10))
term_current=''
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.plot(ts,sawtooth(ts))
    plt.plot(ts,sawtooth_FS(ts,i+1))
    term_current+=terms[i]
    plt.title(rf'$\pi$+{term_current}')
plt.show()
```

```python
# displaying a rectangular wave
def rwave(t,width=2,T=4):
    return np.remainder(t+width/2,T)<width
ts=np.linspace(-8,8,100)
plt.plot(ts,rwave(ts))
ax=plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.axis('equal')
plt.title('A rectangular wave')
plt.show()
```
