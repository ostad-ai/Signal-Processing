### Signal Processing
#### Discrete-time Fourier transform (DTFT):
$X(\omega)=\sum_{n=-\infty}^{\infty}x[n] e^{-j\omega n}$

```python
import numpy as np
from matplotlib import pyplot as plt
from cmath import exp
```

```python
# here, we assume the input x[n]=0 for n<0
def Xdftf(omega,x):
    L=len(x)
    thesum=0.
    for n in range(L):
        thesum+=x[n]*exp(-1j*omega*n)
    return thesum
```

```python
x=[1,1,1,1,1,1,1,1] #the discrete-time signal
omegas=np.linspace(-np.pi,np.pi,100)
X_abs=[]
for n,omega in enumerate(omegas):
    X_abs.append(abs(Xdftf(omega,x)))
plt.plot(omegas,X_abs,'.-')
plt.title('$|X(\omega)|$')
plt.xlabel('$\omega$')
plt.show()
```
