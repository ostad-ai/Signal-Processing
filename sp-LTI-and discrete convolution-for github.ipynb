### Signal Processing
#### Linear Time-Invariant (LTI) system:
$y[n]=(x*h)[n]=\sum_{m}x[m]h[n-m]$

```python
from matplotlib import pyplot as plt
```

```python
#discrete convolution: 1-dimensional
def convolve1D(x,h):
    M=len(h)-1; L=len(x)
    y=[]
    for n in range(L+M):
        conv_sum=0
        for m in range(max(0,n-L+1),min(n,M)+1):
            conv_sum+=h[m]*x[n-m]
        y.append(conv_sum)
    return y
```

```python
# example
h=[.33,.33,.33]
x=[1,5,-5,3,5,8,-6,5]
y=convolve1D(x,h)
print('input signal:',x)
print('impulse response:',h)
print('output signal x*h:',['{:.2f}'.format(elem) for elem in y])
```

```python
# the impulse response here smooths the input signal
plt.plot(x,label='input signal: $x$')
plt.plot(y,label='output signal: $y$')
plt.legend()
plt.title('$y=x*h$')
plt.show()
```
