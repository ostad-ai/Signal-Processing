#### Continuous-time LTI systems: Eigenfunctions and eigenvalues 
$if\; x(t)=e^{j\omega t}, then\; y(t)=H(\omega) x(t)$
<br> such that: $H(\omega)=\int_{-\infty}^{\infty}h(t)e^{-j\omega t} dt$

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# see section of continuous-time Fourier transform (CTFT)
def Hctft(omega,h,ts):
    h_exp=h(ts)*np.exp(-1j*omega*ts)
    return np.trapz(h_exp,ts)  

#approximate continuous-time unit impulse
def delta(t,Delta=1e-1):
    if 0<=t<=Delta:
        return 1./Delta
    else:
        return 0.
    
# impulse response h(t) delays the signal by 5
def h(t):
    if type(t)==np.ndarray: 
        hs=np.zeros_like(ts)
        for i,t in enumerate(ts):
            hs[i]=delta(t-5) 
        return hs
    else:
        return delta(t-5)

#continuous-time complex exponetial e^j.omega.t
def x(omega,t):
    return np.exp(1j*omega*t)
```

```python
freq=.1; omega= 2*np.pi*freq
ts=np.linspace(-10,10,1000)
xs=x(omega,ts)
ys=Hctft(omega,h,ts)*xs
plt.plot(ts,xs.real,label='$x(t)=e^{j\omega t}$ (real part)')
plt.plot(ts,ys.real,label='$y(t)=H(\omega)x(t)$ (real part)')
plt.title('impulse response is: $h(t)=\delta(t-5)$')
plt.legend(); plt.show()
```
