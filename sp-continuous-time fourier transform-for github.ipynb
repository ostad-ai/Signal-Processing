### Signal Processing
#### Continuous-time Fourier transform:
$X(\omega)=\int_{-\infty}^{\infty} x(t) e^{-j\omega t}dt$
<br>
$x(t)=\frac{1}{2\pi}\int_{-\infty}^{\infty}X(\omega)e^{j\omega t}d\omega$

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# x(t)=exp(-at)*u(t)
def x(a,t):
    tsp=np.where(t>=0)
    ts=np.zeros_like(t)
    ts[tsp]=1
    return ts*np.exp(-a*t)
ts=np.linspace(-10,10,100)
a=1.
xs=x(a,ts)
plt.plot(ts,xs)
plt.title('$x(t)=e^{-at}u(t)$')
plt.show()
```

```python
# Computing Fourier transform of x(t)
# Exact versus approximate
omegas=np.linspace(-2*np.pi,2*np.pi,100)
X=[]
for omega in omegas:
    x_exp=x(a,ts)*np.exp(-1j*omega*ts)
    X.append(np.trapz(x_exp,ts))
X=np.asarray(X)
plt.plot(omegas,abs(1/(a+1j*omegas)),'-.',label='Exact Fourier transform')
plt.plot(omegas,abs(X),'-.',label='Approximate Fourier transform')
plt.legend();plt.title(r'$\mathcal{F}\{e^{-at}u(t)\}=\frac{1}{a+j\omega}$')
plt.show()
```
