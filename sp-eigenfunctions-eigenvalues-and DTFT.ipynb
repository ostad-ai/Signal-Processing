```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# here, we assume the input x[n]=0 for n<0
# see section of discrete-time Fourier transform (DTFT)
def Xdtft(omega,x):
    L=len(x)
    thesum=0.
    for n in range(L):
        thesum+=x[n]*np.exp(-1j*omega*n)
    return thesum

#discrete-time unit impulse
def delta(n):
    if n==0:
        return 1.
    else:
        return 0
```

```python
def sine(omega,n):
    return .5*(np.exp(1j*omega*n)+np.exp(-1j*omega*n))
#impulse response: here, h[n] smooths the input signal
def h(n):
    return (delta(n)+delta(n-1)+delta(n-2))/3.
h_n=[h(0),h(1),h(2)] 
freq=.05
omega= 2*np.pi*freq
n_max=100
x_n=[]; y_n=[]
for n in range(n_max):
    temp=sine(omega,n)+sine(omega/3,n)
    x_n.append(temp.real)
    y_n.append((Xdtft(omega,h_n)*temp).real)
plt.plot(y_n,'.-',label='$y[n]=\sum_{k}c_k H(\omega_k)e^{j\omega_k n}$')
plt.plot(x_n,'.-',label='$x[n]=\sum_{k}c_ke^{j\omega_k n}$')
plt.legend(); plt.show()
```
